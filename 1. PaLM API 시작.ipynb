```python
import pandas as pd
import seaborn as sns
from IPython.display import Markdown, display
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from vertexai.language_models import TextGenerationModel, \
                                     TextEmbeddingModel, \
                                     ChatModel, \
                                     InputOutputTextPair, \
                                     CodeGenerationModel, \
                                     CodeChatModel
```

```python
generation_model = TextGenerationModel.from_pretrained("text-bison@001")
```

```python
prompt = "Large Language Model 이 무엇인가?"
response = generation_model.predict(prompt=prompt)
print(response)
```

```python
prompt2 = """
기술산업 최신 트랜드에 대한 5단어 미만으로 10개의 항목을 만들어줘.
"""
response2 = generation_model.predict(prompt=prompt2)
print(response2)
```

## parameters : termperature / max_output_tokens / top_p / top_k

 - temperature: 0.0 - 1.0 (default: 0.0)
 - max_output_tokens: (range: 1 - 1024, default 128)
 - top_p: 0.0 - 1.0, 0.95

```python
prompt3 = """문장을 완성하시오. 
문장: 오늘 아침 나는 내 몸이 갈기갈기 찟어지는 느낌으로 침대에서 일어났다. 어제 헤어진 여자친구는 
"""
response3 = generation_model.predict(prompt=prompt3, temperature=0.1, max_output_tokens=1024, top_p=0.1)
print(response3.text)
```

## chat-bison

```python
chat2 = chat_model.start_chat(
    context="내 이름은 케빈이고 너는 개인 비서야. 내가 가장 좋아하는 영화는 반지의 제왕이이야.",
    examples=[
        InputOutputTextPair(
            input_text="너는 누구를 위해 일하지?",
            output_text="케빈을 위해 일하죠.",
        ),
        InputOutputTextPair(
            input_text="내가 좋아하는 것은?",
            output_text="당신은 영화를 좋아해요.",
        ),
    ],
    temperature=0.3,
    max_output_tokens=200,
    top_p=0.8,
    top_k=40,
)
print(chat.send_message("내가 가장 좋아하는 책은 무엇일까??"))
```

## textembedding-gecko

```python
embedding_model = TextEmbeddingModel.from_pretrained("textembedding-gecko@001")
embeddings = embedding_model.get_embeddings(["인생은 뭐지?"])

for embedding in embeddings:
    vector = embedding.values
    print(f"length: {len(vector)}")
    print(vector)
```

```python
text = [
    "날개야 다시 돋아라.",
    "날자. 날자. 날자. 한 번만 더 날자꾸나.",
    "한 번만 더 날아 보자꾸나.",
    "나는 피로와 공복 때문에 뚫어저 드러 가는 몸동이를 끌고 그 회탁의 거리 속으로 섞겨 들어가지 않을 수도 없다 생각하였다.",
    "그러나 나는 이 발길이 아내에게로 도라가야 옳은가 이것만은 분간하기가 좀 어려웠다."
]
text_df = pd.DataFrame(text, columns=["text"])
text_df
```

```python
text_df["embeddings"] = [
    emb.values for emb in embedding_model.get_embeddings(text_df.text.values)
]
text_df
```

## cosine similarity로 text 예시의 유사성 비교

```python
cos_sim_array = cosine_similarity(list(text_df.embeddings.values))
sim_df = pd.DataFrame(cos_sim_array, index=text, columns=text)
sim_df
```
